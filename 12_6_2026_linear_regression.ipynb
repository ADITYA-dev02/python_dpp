```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data={'Studyhours':[2,3,4,5],
      'marks':[40,50,60,70]}
df= pd.DataFrame(data)
x=df[['Studyhours']]
y=df['marks']
model=LinearRegression()
model.fit(x,y)
prediction = model.predict([[8.2]])
print(prediction)
```

```python
#question 1 and 2 read aand print
import pandas as pd

df=pd.read_csv("student.csv")
print(df.info())
```

```python
#question 3 identify misssing values
print(df.isnull())
```

```python
# question 4 5 replace missign values
df['Marks'].fillna(int(df['Marks'].mean()),  inplace=True)
df['Attendance'].fillna(int(df['Attendance'].mean()),  inplace=True)
print(df)
```

```python
#question 6 drop the name column
df.drop('Name', axis=1, inplace=True)
print(df)
```

```python
#question 7 missing values
print(df.isnull())
```

```python
#question 8 marks as target variable and attendace aas input
from sklearn.linear_model import LinearRegression

x=df[['Attendance']]
y=df['Marks']
model=LinearRegression()
model.fit(x,y)
a=int(input("enter attendance for prediction : "))
prediction = model.predict([[a]])
print("predicted marks for ",a,"% attendance is :",prediction)
```
